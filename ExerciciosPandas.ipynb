1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
import pandas as pd
df_notas=pd.DataFrame(dados_alunos.drop(columns=['Nome','Frequencia','RA']))
media=[]
for i in df_notas.index:
    media.append(df_notas.loc[i].mean())
dados_alunos['Media']=media
dados_alunos


```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
import pandas as pd 
maximo=dados_alunos['Media'].idxmax()
min=dados_alunos['Media'].idxmin()
print('O aluno com maior média foi',dados_alunos.loc[maximo]['Nome'])
print('O aluno com menor média foi',dados_alunos.loc[min]['Nome'])
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
import pandas as pd
cadastro=pd.read_excel('cadastro_alunos.xlsx')
dados_totais=pd.merge(dados_alunos,cadastro,how='outer')
dados_totais
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
import pandas as pd
dados_totais
print('A média do público feminino é:', dados_totais.query('Sexo=="Feminino"')['Media'].mean())
print('A média do público masculino é:', dados_totais.query('Sexo=="Masculino"')['Media'].mean())
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
import pandas as pd
print('A média de idade de quem tirou acima de 7:',  dados_totais.query('Media>=7')['Idade'].mean())
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
import pandas as pd
cidade_maior_media=dados_totais.groupby('Cidade').mean(numeric_only=True)['Media'].idxmax()
media_maior_por_cidade=dados_totais.groupby('Cidade').mean(numeric_only=True)['Media'].max()
print(f'A cidade com maior média foi {cidade_maior_media} com média de {media_maior_por_cidade}')
```

7. Leia o arquivo `aluno3.csv`, apenas as colunas RA, Prova_1, Prova_2, Prova_3, Prova_4 calcule a médias das notas da prova e salve em um arquivo excel (xlsx) as colunas RA e Média.


```python
import pandas as pd
alunos_3=pd.read_csv('alunos3.csv',usecols=[0,3,4,5,6],sep=';',decimal=',')
media_notas=[]
for i in alunos_3.index:
    media_notas.append(alunos_3.drop(columns='RA').loc[i].mean())
alunos_3['Media_Notas']=media_notas
alunos_3.to_excel('exercicio_7.xlsx')
```
